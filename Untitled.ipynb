```python
df_merged = pd.read_excel("df_full.xlsx")
```

```python
df_merged.head(5)
```

```python
df_pos_each_year = df_merged[df_merged.Rnd < 4.0].groupby(['Year', 'Tm','Pos_x']).size().reset_index(name='cnt')
```

```python
teams = list(df_pos_each_year.Tm.unique())
for t in teams:
    temp_df = df_pos_each_year[df_pos_each_year.Tm == t]
    fig = px.line(temp_df, x="Year", y="cnt", color='Pos_x', title = "Team : " + t)
    fig.show()
```
